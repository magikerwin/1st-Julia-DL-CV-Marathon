## Day 007 作業：使用條件與迴圈找出數值

運用迴圈及條件判斷，找出並印出範圍內 (例如從 1 到 100)，是 3 的倍數，但是無法被 5 整除的數字。

````julia
x = 1 #範圍起始
y = 100 #範圍結束
````

````julia
# 程式碼
for i in x:y
    if (i%3==0) && (i%5!=0)
        println(i)
    end
end
````
