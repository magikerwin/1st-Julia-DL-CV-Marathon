### 作業：範例程式示範了 Int64 及 Float64 的最小值和最大值，作業請列出並觀察其他整數及浮點數型別的最小值和最大值。

呼叫 `typemin()` 及 `typemax()` 函式得到各型別的最小值和最大值。

列出 Int128, Int32, Int16, Int8, UInt128, UInt64, UInt32, UInt16, UInt8 整數型別的最小值及最大值

```julia
len = 40
println("$(rpad(UInt128,7)): min $(rpad(typemin(Int128), len)), max $(typemax(Int128))")
println("$(rpad(Int64,7)): min $(rpad(typemin(Int64), len)), max $(typemax(Int64))")
println("$(rpad(Int32,7)): min $(rpad(typemin(Int32), len)), max $(typemax(Int32))")
println("$(rpad(Int16,7)): min $(rpad(typemin(Int16), len)), max $(typemax(Int16))")
println("$(rpad(Int8,7)): min $(rpad(typemin(Int8), len)), max $(typemax(Int8))")
println()
println("$(rpad(UInt128,7)): min $(rpad(typemin(UInt128), len)), max $(typemax(UInt128))")
println("$(rpad(UInt64,7)): min $(rpad(typemin(UInt64), len)), max $(typemax(UInt64))")
println("$(rpad(UInt32,7)): min $(rpad(typemin(UInt32), len)), max $(typemax(UInt32))")
println("$(rpad(UInt16,7)): min $(rpad(typemin(UInt16), len)), max $(typemax(UInt16))")
println("$(rpad(UInt8,7)): min $(rpad(typemin(UInt8), len)), max $(typemax(UInt8))")
```

列出 Float32, Float16 浮點數型別的最小和最大值

```julia
len = 5
println("$(rpad(Float64,7)): min $(rpad(typemin(Float64), len)), max $(typemax(Float64))")
println("$(rpad(Float32,7)): min $(rpad(typemin(Float32), len)), max $(typemax(Float32))")
```
